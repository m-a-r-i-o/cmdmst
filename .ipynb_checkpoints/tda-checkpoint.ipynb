```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import kmapper as km
#from sklearn.preprocessing import RobustScaler
from kmapper.jupyter import display
```

```python
CMD = pd.read_csv('prova.csv')[["V1", "V2"]]
CMD.columns = ["Mag", "Col"]

CMD_2pop = pd.read_csv('prova_2pop.csv')[["V1", "V2"]]
CMD_2pop.columns = ["Mag", "Col"]
```

```python
def plot_CMD(CMD):
    plt.scatter(CMD["Col"], -CMD["Mag"], s = 1, color="#18A4A6")
    plt.show()
    
def kmapper_CMD(CMD):
    mapper = km.KeplerMapper(verbose=1)
    data = CMD.to_numpy()
    projected_data = mapper.fit_transform(data, projection=[0,1]) # X-Y axis
    cover = km.Cover(n_cubes=4)
    graph = mapper.map(projected_data, data, cover=cover)
    _ = mapper.visualize(graph, path_html="keplermapper_output.html",
                 title="CMD")
    display("keplermapper_output.html")
    return graph
```

```python
plot_CMD(CMD)
graph_1pop = kmapper_CMD(CMD)
```

```python
plot_CMD(CMD_2pop)
graph_2pop = kmapper_CMD(CMD_2pop)
```

```python
def featurize_graph(graph):
    mean_edges_per_vertex = np.array([len(item[1]) for item in graph['links'].items()]).mean()
    median_edges_per_vertex = np.median(np.array([len(item[1]) for item in graph['links'].items()]))
    max_edges_per_vertex = np.array([len(item[1]) for item in graph['links'].items()]).max()
    min_edges_per_vertex = np.array([len(item[1]) for item in graph['links'].items()]).min()
    num_vertices = len(graph['nodes'])
    return num_vertices, min_edges_per_vertex, mean_edges_per_vertex, median_edges_per_vertex, max_edges_per_vertex
```

```python
featurize_graph(graph_1pop)
```

```python
featurize_graph(graph_2pop)
```
